## 2. Q & A

### 10. Produce a list of facilities with a total revenue less than 1000. The output of facility name and total revenue, sorted by revenue. Remember that there's a different cost for guests and members.

```python
sql_query10 = """
SELECT facility, SUM(FINALCOST) as totalrev
FROM(
    SELECT FACILITIES.name as facility,
    CASE WHEN BOOKINGS.memid = 0 
        THEN FACILITIES.guestcost * BOOKINGS.slots
        ELSE FACILITIES.membercost * BOOKINGS.slots
        END AS finalcost
        FROM BOOKINGS
    INNER JOIN FACILITIES ON FACILITIES.facid = BOOKINGS.facid
    INNER JOIN MEMBERS on MEMBERS.memid = BOOKINGS.memid) sub
GROUP BY facility
HAVING totalrev < 1000
ORDER BY totalrev;
"""
for row in cursor.execute(sql_query10):
    print(row)
```

### 11. Produce a report of members and who recommended them in alphabetic surname,firstname order.

```python
sql_query11 = """
WITH rec_by_list AS (SELECT memid, firstname AS rec_by_firstname, surname AS rec_by_surname FROM MEMBERS)
SELECT MEMBERS.firstname, MEMBERS.surname, rec_by_list.rec_by_firstname, rec_by_list.rec_by_surname
FROM MEMBERS
LEFT JOIN rec_by_list ON MEMBERS.recommendedby = rec_by_list.memid
ORDER BY surname, firstname
"""

for row in cursor.execute(sql_query11):
    print(row)
```

### 12. Find the facilities with their usage by member, but not guests.

```python
sql_query12 = """
SELECT FACILITIES.name, COUNT(BOOKINGS.bookid) AS times_booked, MEMBERS.firstname, MEMBERS.surname
FROM BOOKINGS
INNER JOIN FACILITIES ON BOOKINGS.facid = FACILITIES.facid
AND BOOKINGS.memid > 0
INNER JOIN MEMBERS ON BOOKINGS.memid = MEMBERS.memid
GROUP BY FACILITIES.name, MEMBERS.firstname, MEMBERS.surname
"""

for row in cursor.execute(sql_query12):
    print(row)
```

### 13. Find the facilities usage by month, but not guests.

```python
sql_query13 = """
SELECT FACILITIES.name, COUNT(BOOKINGS.bookid) AS times_booked, strftime('%m', BOOKINGS.starttime) as month
FROM BOOKINGS
INNER JOIN FACILITIES ON BOOKINGS.facid = FACILITIES.facid
AND BOOKINGS.memid > 0
GROUP BY FACILITIES.name, month
"""

for row in cursor.execute(sql_query13):
    print(row)
```
